```python
# Python program to print all heading tags
import pickle
from IPython import display
import math
from pprint import pprint
import requests
import nltk
from bs4 import BeautifulSoup
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import random
from wordcloud import WordCloud
import os
import spacy
from textblob import TextBlob
from pattern.en import sentiment
import urllib3 
import re
import datetime
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
nlp = spacy.load('en_core_web_sm')
%matplotlib inline
sns.set(style='darkgrid', context='talk', palette='Dark2')

cleanData = []

utc_datetime = datetime.datetime.utcnow()
formated_string = utc_datetime.strftime("%Y-%m-%d-%H%MZ")
filename = 'Scraped_%s.txt'% formated_string


# scrape a chosen URL
url_link = 'https://www.theguardian.com/us'
request = requests.get(url_link)

Soup = BeautifulSoup(request.text, 'lxml')
    
tag_words =[]
# creating a list by type of heading tags
heading_tags = [ "h3"] #the user will have to determine what "h" tags are valid for given URL
for tags in Soup.find_all(filename):
    tag_words.append(tags.text.strip())
    list2 = [x.replace('\n', ' ') for x in tag_words]
    list3 = [strg.lower() for strg in list2]
    

# with open(filename,'w') as h_file: #if you want to save a text copy to avoid utf-8 error message
#     h_file.write('\n'.join(list3))
    

    
regex = re.compile('[^a-zA-Z]')
for i in list3: #slice based on data you want to include. Some junk data can be removed
    if len(i)>18:
        print(i, len(i))
        cleanData.append(regex.sub(' ', i))

with open(filename, "wb") as fp:   #Pickling
    pickle.dump(cleanData, fp)
```
